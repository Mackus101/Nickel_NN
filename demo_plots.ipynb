```python
import numpy as np
import matplotlib.pyplot as plt

def ReLU(x):
    return x * (x > 0)

test_space = np.linspace(-4,4,100)

y = ReLU(test_space)
```

```python
plt.plot(test_space, y)
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.savefig("pictures/ReLU.pdf", bbox_inches='tight')
```

```python
y_2 = np.tanh(test_space)
```

```python
plt.plot(test_space, y_2)
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.savefig("pictures/tanh.pdf", bbox_inches='tight')
```
